```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df
```

## Player Count

* Display the total number of players


```python
#total number of players using int value for calculating value - shows 576 players

sn_player = len(purchase_data_df["SN"].value_counts())
sn_player_count = pd.DataFrame([sn_player], columns = ["Total Players"])
sn_player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
## all of these you will probably have to len over to get the data correctly

#get unique values - item name is an object
unique_items = purchase_data_df["Item Name"].unique()
#unique_items #listed every unique item name

#count - 
count_item = len(purchase_data_df["Item Name"].value_counts())
count_item
#provides a value of 179

#get average purchase price - float
price_average = purchase_data_df["Price"].mean()
price_average
#gives result 3.050987179487176


#get total number of purchases- float
purchase_count = purchase_data_df["Purchase ID"].count()
purchase_count
#gives result of 780

#get total revenue - "price" is a float
revenue_total = purchase_data_df["Price"].sum()
revenue_total
#gives result of 2379.77


#average total per person


#Purchasing Analysis dictionary to dataframe 
purchase_dicts = ({"Number of Unique Items": [count_item],
                        "Average Purchase Price": [price_average],
                        "Total Number of Purchases": [purchase_count],
                        "Total Revenue": [revenue_total]})
pur_analysis_df = pd.DataFrame(purchase_dicts)
pur_analysis_df.style.format({"Average Purchase Price": "${0:.2f}",
                             "Total Revenue":"${0:.2f}"})     

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Group by gender and SN
groupby_gender = purchase_data_df.groupby("Gender")
#groupby_gender


#Arrange for counts and means(113,652, 15)

gender_purchase_count=groupby_gender["Purchase ID"].count()
gender_purchase_count

#price_average(3.20, 3.02, 3.35)
gender_price_av=groupby_gender["Price"].mean()
gender_price_av

#purchase_sum(361.94, 1967.64, 50.19)
gender_sum=groupby_gender["Price"].sum()
gender_sum

groupby_sn = purchase_data_df.groupby("SN")
unique_sn = groupby_gender["SN"].nunique()
#sn_sum
sn_sum = groupby_sn["Price"].sum
sn_sum

##sn average (4.47, 4.07, 4.56)
sn_average=(groupby_gender["Price"].sum()/unique_sn)
sn_average


#build DF for display
purchase_analysis_df = pd.DataFrame({"Purchase Count": gender_purchase_count, 
                                    "Average Purchase Price": gender_price_av,
                                    "Total Purchase Value":gender_sum,
                                    "Avg Purchase Total per Person": sn_average})

purchase_analysis_df["Average Purchase Price"].astype(float)
purchase_analysis_df["Total Purchase Value"].astype(float)
#Set index as gender
purchase_analysis_df.index.name = "Gender"

#format DF for 
purchase_analysis_df.style.format({"Total Purchase Value":"${0:.2f}",
                                  "Average Purchase Price":"${0:.2f}",
                                  "Avg Purchase Total per Person":"${0:.2f}"})

```

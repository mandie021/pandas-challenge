```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
## all of these you will probably have to len over to get the data correctly

#get unique values - item name is an object
unique_items = purchase_data_df["Item Name"].unique()
#unique_items #listed every unique item name

#count - 
count_item = len(purchase_data_df["Item Name"].value_counts())
count_item
#provides a value of 179

#get average purchase price - float
price_average = purchase_data_df["Price"].mean()
price_average
#gives result 3.050987179487176


#get total number of purchases- float
purchase_count = purchase_data_df["Purchase ID"].count()
purchase_count
#gives result of 780

#get total revenue - "price" is a float
revenue_total = purchase_data_df["Price"].sum()
revenue_total
#gives result of 2379.77


#average total per person


#Purchasing Analysis dictionary to dataframe 
purchase_dicts = ({"Number of Unique Items": [count_item],
                        "Average Purchase Price": [price_average],
                        "Total Number of Purchases": [purchase_count],
                        "Total Revenue": [revenue_total]})
pur_analysis_df = pd.DataFrame(purchase_dicts)
pur_analysis_df.style.format({"Average Purchase Price": "${0:.2f}",
                             "Total Revenue":"${0:.2f}"})     

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group by SN
spenders_sn = purchase_data_df.groupby("SN")

#count purchase ID's -int
spender_purchase = spenders_sn["Purchase ID"].count()

#mean the price by SN - float 
spender_average = spenders_sn["Price"].mean()
spender_average

#sum totals by SN - float
spender_total = spenders_sn["Price"].sum()
spender_total

#df of values
top_spender = pd.DataFrame({"Purchase Count": spender_purchase,
                             "Average Purchase Price": spender_average,
                             "Total Purchase Value":spender_total})

#sort in decending order
formatted_spender = top_spender.sort_values(["Total Purchase Value"], ascending=False).head()

#format with currency"${:,.2f}"
formatted_spender.style.format({"Average Purchase Total":"${:.2f}",
                                 "Average Purchase Price":"${:.2f}", 
                                 "Total Purchase Value":"${:.2f}"})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create df with Item ID, Item Name, and Price
pop_items = purchase_data_df [["Item ID", "Item Name", "Price"]]
pop_items

#group by id and name
items_groups = pop_items.groupby(["Item ID", "Item Name"])

#count of items -780
items_count = items_groups["Price"].count()
items_count

#sum of item - 2379.77
item_value = items_groups["Price"].sum()
item_value

#value of individual item - 3.05
item_price = item_value/items_count
item_price

#df of values
most_popular_items = pd.DataFrame({"Purchase Count": items_count, 
                                   "Item Price": item_price,
                                   "Total Purchase Value":item_value})

#sort items by purchase count
most_popular_formatted = most_popular_items.sort_values(["Purchase Count"], ascending = False).head()

#format sf "${:.2f}" and left alignment
popular_formatted = most_popular_formatted.style.format({"Item Price":"${:.2f}","Total Purchase Value":"${:.2f}"})

popular_formatted
```
